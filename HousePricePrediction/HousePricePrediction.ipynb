```python
#step1 : import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

2 - Load Dataset

```python
#step2 : read the house price dataset and check for null
df=pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/House%20Prices.csv")
```

```python
#Get first 5 rows of dataframe
df.head()
```

```python
#Get info of dataframe
df.info()
```

```python
#get dataframe statistics
df.describe()
```

```python
#Get column names
df.columns
```

```python
#Get shape of dataframe
df.shape
```

```python
y=df['Price']

y.shape
```

```python
y
```

```python
X=df.drop('Price',axis=1)
X.shape
```

```python
X
```

```python
print(X.dtypes)
```

```python
X = X.select_dtypes(include=['int64','float64'])
```

3 - Data Preprocessing

```python
#preprocessing
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
X=mm.fit_transform(X)
X
```

4 - Train-test split

```python
#train test split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape,y_test.shape
```

5 - Train Model

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

6 - Predict

```python
#get model prediction
y_pred=lr.predict(X_test)
y_pred.shape
```

```python
y_pred
```

7 - Model Evaluation

```python
#Model evaluation
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

8 - Get Visualization

```python
#Get visualization
import matplotlib.pyplot as plt
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Prize")
plt.show()
```
